## Cropping Image

```python
blur_crop_path = 'Images/regular_crop_blur/'
noise_crop_path = 'Images/regular_crop_noise/'
crop_path = 'Images/regular_crop/'
# os.mkdir(blur_crop_path)
# os.mkdir(noise_crop_path)
# os.mkdir(crop_path)
def crop_image(image, shift=True):
    a, b = 0,0
    if shift: 
        a = 20
        b = 20
    x, y = 50, 110
    h, w = 325, 325
    crop_img = img[y:y+h+a, x:x+w+b]
    return crop_img


for file in regular_files: 
    path = image_path + '/' + file
    images = os.listdir(path)
    regular_path_card = crop_path + file
    blur_path_card = blur_crop_path + file
    noise_path_card = noise_crop_path + file
    os.mkdir(regular_path_card)
    os.mkdir(blur_path_card)
    os.mkdir(noise_path_card)
    for image in images: 
        
        img = cv2.imread(path + '/' + image)
        img_crop = crop_image(img)
        cv2.imwrite(regular_path_card + '/' + image, img_crop)
        # Blur Image
        cv2.imwrite(blur_path_card + '/' + image, cv2.blur(img_crop, ksize))
        # Noise Images
        # cv2.imwrite(noise_path_card + '/' + image, random_noise(dm, mode='s&p',amount=.5))
    
```

```python
import cv2
import matplotlib.pyplot as plt
orb = cv2.ORB_create()


def calculateHOGpointsSingleImg(orb,img):
  
    assert (img is not None)
    
    kpts, descs = orb.detectAndCompute(img,None)

    return (kpts,descs)
```
